```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from spit.analysis.functions import cell_agg_tracks

plt.style.use(r'U:\DNA-PAINT-SPT\nature_7pt.mplstyle')


def plot_aggregated_tracks_paper(path, start, color_tracks, color_frame='black'):
    df_tracks = pd.read_csv(path)
    df_stats = pd.read_hdf(os.path.splitext(path)[0]+'_stats.hdf', key='df_data')

    df_statsF = df_stats.loc[(df_stats.length > 10)]
    keep_tracks = df_statsF['track.id'].values
    df_tracksF = df_tracks.loc[df_tracks['track.id'].isin(keep_tracks)]

    plt.rc('axes', edgecolor=color_frame)

    # without scalebar
    f, ax = plt.subplots(1, 1)
    f.set_figheight(0.95)
    f.set_figwidth(0.95)
    f.subplots_adjust(left=0.02, right=0.95, bottom=0.05,
                      top=0.95, wspace=0.2, hspace=0.0)
    plt.rc('axes', edgecolor=color_frame)
    cell_agg_tracks.plot_aggregated_tracks(df_tracksF, df_statsF, start=start, duration=250,
                                           color=color_tracks, scalebar=False, dt=0.08, px2nm=108,
                                           path=os.path.dirname(path), ax=ax)

    ax.set_title('')
    ax.set_aspect('equal')
    ax.set_rasterized(True)
    ax.spines.top.set_visible(True)
    ax.spines.right.set_visible(True)


# Paths
path_df_tracks_DNA = r'U:\DNA-PAINT-SPT\Fig2E\df_DNAPAINT.csv'
path_df_tracks_SD = r'U:\DNA-PAINT-SPT\Fig2E\df_SD.csv'
```

```python
plot_aggregated_tracks_paper(path=path_df_tracks_DNA,
                             start=0, color_tracks=None, color_frame='tomato')
plot_aggregated_tracks_paper(path=path_df_tracks_DNA,
                             start=4500, color_tracks=None, color_frame='tomato')

plot_aggregated_tracks_paper(path=path_df_tracks_SD,
                             start=0, color_tracks=None,  color_frame='cornflowerblue')
plot_aggregated_tracks_paper(path=path_df_tracks_SD,
                             start=4500, color_tracks=None,  color_frame='cornflowerblue')
```
