```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use(
    r'U:\DNA-PAINT-SPT\nature_7pt.mplstyle')

# Load data
data = pd.read_hdf(r'U:\DNA-PAINT-SPT\SI\Fig11\data.hdf')
```

```python
# Plot
f, ax = plt.subplots(1, 1)
f.set_figheight(1.2)
f.set_figwidth(4)

sns.violinplot(data=data, x='surface', y='ratio_density', hue='channel', split=True,
               palette=['orchid', 'cyan'], ax=ax)

# Decorate plot
ax.set_ylabel('Localizations/Background \n(area normalized) ')
ax.set(xlabel=None)
ax.tick_params(axis='x', rotation=30)
leg = ax.legend(loc='upper left', handles=ax.legend_.legendHandles, labels=[
                'ATTO643', 'Cy3B'], handlelength=1, borderpad=0.2, labelspacing=0.2)
leg.get_frame().set_edgecolor('black')
leg.get_frame().set_linewidth(0.5)
```
